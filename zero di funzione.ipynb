```python
from math import exp
from numpy import abs
import matplotlib.pyplot as plt
```

```python
def f(x):
  return exp(x) - x**2
```

```python
f = lambda x: exp(x) - x**2
```

```python
x = [-5 + i * 0.1 for i in range(101)]
y = [f(e) for e in x]
plt.plot(x, y)
```

```python
def sec_step(fun, x1, x0):
  return x1 - (x1 - x0)/(fun(x1) - fun(x0)) * fun(x1)
```

```python
xa = -5
xb = 5
```

```python
sec_step(f, xa, xb)
```

```python
def sec_step_arr(fun, arr):
  return sec_step(fun, arr[-1], arr[-2])
```

```python
x_sol = [xa, xb]
# x_sol += [sec_step_arr(f, x_sol)]
epsilon = 1e-6
while abs(x_sol[-1] - x_sol[-2]) > epsilon:
  x_sol += [sec_step_arr(f, x_sol)]
print(x_sol)
```

```python
y_sol = [f(e) for e in x_sol]
```

```python
x = [-5 + i * 0.1 for i in range(101)]
y = [f(e) for e in x]
plt.plot(x, y)
plt.plot(x_sol, y_sol, '-o')
```
