# Avvolgitore

## Obiettivo
Determinare la coppia di accelerazione della bobina per portarla in rotazione alla velocità data, svolgendo una quantità data di film.

## Inizializzazione

```python
import numpy as np
from math import pi
import matplotlib.pyplot as  plt
```

```python
mm = 0.001
min = 60 * 1
rpm = 2 * pi / min
```

## Dati

```python
d_i = 76 * mm
d_e_max = 500 * mm
l_reel = 800 * mm
rho_reel = 2750 * 0.95
omega_end = 300 * rpm
s_unwind = 1200 * mm
C_lim = 150
```

## Svolgimento calcoli

Si ipotizza svolgimento ad accelerazione costante

```python
# diameter = 300 * mm
diameter = np.linspace(d_i, d_e_max, num=100)
```

```python
radius = 0.5 * diameter
v_end = omega_end * (radius)
delta_t = 2 * s_unwind / v_end
acc = 2 * s_unwind / delta_t**2
acc_ang = acc/radius
```

```python
I_z = 0.5 * pi * (radius**4 - (0.5*d_i)**4) * l_reel * rho_reel
```

```python
C_unw = acc_ang * I_z
```

```python
plt.plot(diameter, C_unw)
```

```python
power_max = C_unw * omega_end
```

```python
plt.plot(diameter, power_max/1000)
```
